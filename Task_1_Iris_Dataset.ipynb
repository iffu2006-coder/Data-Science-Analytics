# Task 1: Exploring and Visualizing Iris Dataset
In this task, I explored and visualized the Iris dataset using Python. The dataset contains flower measurements such as sepal length, sepal width, petal length, and petal width.

## Importing Libraries
In this step, I imported the required Python libraries. Pandas is used for loading and analyzing the dataset, while Matplotlib and Seaborn are used for creating visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset
The Iris dataset was loaded using pandas. The first five rows of the dataset are displayed using the head() function to get an initial view of the data.

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df.head()
```

The table shows the first five records of the Iris dataset. It includes columns such as sepal_length, sepal_width, petal_length, petal_width, and species.

## Dataset Shape
The shape of the dataset shows the number of rows and columns present in the dataset.

```python
print("Shape of dataset:", df.shape)
print("Columns:", df.columns)
df.head()
```

The dataset contains 150 rows and 5 columns. This means there are 150 flower records and 5 features in the dataset.

## Statistical Summary
The describe() function gives a statistical summary of the numerical columns in the dataset.

```python
df.describe()
```

This table shows count, mean, standard deviation, minimum, maximum, and quartile values for numerical columns. It helps us understand the range and spread of the flower measurements.

## Checking Missing Values
Missing values are checked to see if any column has empty or null values.

```python
df.isnull().sum()
```

The result shows that there are no missing values in the dataset. Therefore, no missing value handling is required for this task.

## Scatter Plot
This graph shows the relationship between petal length and petal width. Different colors are used to represent different iris species.

```python
# Scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
plt.title("Scatter Plot of Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

The scatter plot shows that petal length and petal width are strongly related. It also shows that different iris species can be separated clearly using petal measurements.

## Histogram
This histogram shows the distribution of sepal length values in the Iris dataset.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["sepal_length"], bins=20, kde=True)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows how frequently different sepal length values appear in the dataset. Most flowers have sepal length values within a common middle range.

## Box Plot
This box plot shows the spread of sepal length values for each iris species.

```python
# Box plot
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="species", y="sepal_length")
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.show()
```

The box plot helps compare the median, range, and spread of sepal length among different species. It also helps identify possible outliers in the data.

## Conclusion

In this task, the Iris dataset was loaded and explored using pandas. The structure of the dataset was checked using shape, columns, head, info, and describe functions. Missing values were also checked, and no missing values were found. Different visualizations such as scatter plot, histogram, box plot, and pair plot were created using Matplotlib and Seaborn. The analysis shows that petal length and petal width are important features for distinguishing different iris species.
